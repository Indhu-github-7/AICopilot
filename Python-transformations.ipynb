```python
gender = ['M','m','f','','F','x','f','m']
print(gender)
```

```python
for i in range(len(gender)):
  g = gender[i]
  print(g)
  if g.lower()=='m':
    gender[i] = 'Male'
  elif g.lower()=='f':
    gender[i] = 'Female'
  else:
    gender[i] = 'Invalid'
print(gender)
```

```python
dept = [11,12,11,13,12,11,13,11]
sal = [10000,20000,30000,20000,50000,20000,10000,30000]
```

```python
info = list(zip(gender,dept,sal))
print(info)
```

```python
dcnt = {}
for g,d,s in info:
  if dcnt.get(d)==None:
    dcnt[d] = 1
  else:
    dcnt[d] += 1
print(dcnt)
```

```python
dmaxs = {}
for g,d,s in info:
  if dmaxs.get(d)==None:
    dmaxs[d] = s
  else:
    if dmaxs[d]<s:
      dmaxs[d] = s
print(dmaxs)
```

```python
dmins = {}
for g,d,s in info:
  if dmins.get(d)==None:
    dmins[d] = s
  elif dmins[d] > s:
    dmins[d] = s
print(dmins)
```
