```python
gender = ['M','m','f','','F','x','f','m']
len(gender)
```

```python
for i in range(len(gender)):
  g = gender[i]
  if g.lower()=='m':
    gender[i] = 'Male'
  elif g.lower()=='f':
    gender[i] = 'Female'
  else:
    gender[i] = 'Invalid'
print(gender)
```

```python
qual = {101: [('btech', 'nu', 2015, 80),
  ('mtech', 'jntu', 2017, 80),
  ('phd', 'ou', 2022, 90)],
 102: [('bsc', 'ou', 2020, 90), ('mba', 'jntu', 2022, 80)],
 103: [('bcom', 'ou', 2020, 90)]}
```

```python
marks = {}
for key in qual:
  lst = qual[key]
  for item in lst:
    if marks.get(key)==None:
      marks[key] = [item[-1]]
    else:
      marks[key].append(item[-1])
print(marks)
```
